```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
print(df.head())
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.mean())) #filling missing values in Review Rating column with mean of that category
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower() #converting all column names to lower case
df.columns = df.columns.str.replace(' ', '_') #replacing spaces in column names with underscore
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}) #renaming a specific column
```

```python
df = df.drop('promo_code_used', axis=1,)#dropping the promo_code_used column as it is same as discount_applied column
df.head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Weekly': 7,
    'Every 3 months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping) #mapping the purchase_frequency column to a new column purchase_frequency_days with corresponding days
```
